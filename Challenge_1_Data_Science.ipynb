##**Analisis y Visualizacion de datos**

###**1. Análisis de facturación**



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total para Tienda 1: {ingreso_total_tienda1:,.2f}")
print(f"Ingreso total para Tienda 2: {ingreso_total_tienda2:,.2f}")
print(f"Ingreso total para Tienda 3: {ingreso_total_tienda3:,.2f}")
print(f"Ingreso total para Tienda 4: {ingreso_total_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

ingresos_totales = {
    'Tienda 1': ingreso_total_tienda1,
    'Tienda 2': ingreso_total_tienda2,
    'Tienda 3': ingreso_total_tienda3,
    'Tienda 4': ingreso_total_tienda4
}

ingresos_df = pd.Series(ingresos_totales, name='Ingreso Total').reset_index()
ingresos_df.columns = ['Tienda', 'Ingreso Total']

plt.figure(figsize=(16, 8)) # Ajustar el tamaño para barras horizontales
sns.barplot(x='Ingreso Total', y='Tienda', data=ingresos_df, palette='viridis') # Intercambiar x e y
plt.title('Ingreso Total por Tienda')
plt.xlabel('Ingreso Total')
plt.ylabel('Tienda')
plt.ticklabel_format(style='plain', axis='x') # Desactivar notación científica para el eje X
plt.grid(axis='x', linestyle='--', alpha=0.7) # Ajustar la rejilla al eje X
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combinar todos los dataframes en uno solo
all_tiendas = pd.concat([
    tienda.assign(Tienda='Tienda 1'),
    tienda2.assign(Tienda='Tienda 2'),
    tienda3.assign(Tienda='Tienda 3'),
    tienda4.assign(Tienda='Tienda 4')
])

# Convertir 'Fecha de Compra' a formato de fecha
all_tiendas['Fecha de Compra'] = pd.to_datetime(all_tiendas['Fecha de Compra'], format='%d/%m/%Y')

# Extraer el mes y año para agrupar las ventas mensuales
all_tiendas['Mes_Año'] = all_tiendas['Fecha de Compra'].dt.to_period('M')

# Calcular las ventas mensuales para cada tienda
ventas_mensuales = all_tiendas.groupby(['Tienda', 'Mes_Año'])['Precio'].sum().reset_index()

# Convertir 'Mes_Año' de Period a Timestamp para el eje X en el gráfico
ventas_mensuales['Mes_Año'] = ventas_mensuales['Mes_Año'].dt.to_timestamp()

# Crear el lienzo con subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(16,8))
axes = axes.flatten() # Para poder iterar sobre ellos fácilmente

# Obtener la lista de nombres de tiendas
nombres_tiendas = all_tiendas['Tienda'].unique()

# Iterar sobre cada tienda y crear un gráfico de líneas en un subplot diferente
for i, store_name in enumerate(nombres_tiendas):
    store_data = ventas_mensuales[ventas_mensuales['Tienda'] == store_name]
    sns.lineplot(x='Mes_Año', y='Precio', data=store_data, marker='o', ax=axes[i])
    axes[i].set_title(f'Ventas Mensuales - {store_name}')
    axes[i].set_xlabel('Fecha')
    axes[i].set_ylabel('Ingreso Total Mensual')
    axes[i].grid(True, linestyle='--', alpha=0.6)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].ticklabel_format(style='plain', axis='y') # Desactivar notación científica en el eje Y

plt.tight_layout()
plt.show()
```

### **2. Ventas por categoría**

```python
print('Ventas por categoría de la Tienda 1:')
print(tienda['Categoría del Producto'].value_counts())
print('\n')

print('Ventas por categoría de la Tienda 2:')
print(tienda2['Categoría del Producto'].value_counts())
print('\n')

print('Ventas por categoría de la Tienda 3:')
print(tienda3['Categoría del Producto'].value_counts())
print('\n')

print('Ventas por categoría de la Tienda 4:')
print(tienda4['Categoría del Producto'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Función para generar gráficos de ventas por categoría en un subplot específico
def plot_ventas_por_categoria(df, store_name, ax):
    ventas_categoria = df['Categoría del Producto'].value_counts().reset_index()
    ventas_categoria.columns = ['Categoría del Producto', 'Cantidad de Ventas']
    sns.barplot(x='Categoría del Producto', y='Cantidad de Ventas', data=ventas_categoria, palette='pastel', ax=ax)
    ax.set_title(f'Ventas por Categoría en {store_name}')
    ax.set_xlabel('Categoría del Producto')
    ax.set_ylabel('Cantidad de Ventas')
    ax.tick_params(axis='x', rotation=45)

# Crear el lienzo con subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(16, 8))
axes = axes.flatten() # Para poder iterar sobre ellos fácilmente

# Generar gráficos para cada tienda en su subplot correspondiente
plot_ventas_por_categoria(tienda, 'Tienda 1', axes[0])
plot_ventas_por_categoria(tienda2, 'Tienda 2', axes[1])
plot_ventas_por_categoria(tienda3, 'Tienda 3', axes[2])
plot_ventas_por_categoria(tienda4, 'Tienda 4', axes[3])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_calificacion_distribution(df, store_name, ax):
    sns.countplot(x='Calificación', data=df, palette='viridis', ax=ax)
    ax.set_title(f'Distribución de Calificaciones - {store_name}')
    ax.set_xlabel('Calificación')
    ax.set_ylabel('Número de Evaluaciones')
    ax.set_ylim(0, df['Calificación'].value_counts().max() * 1.1) # Ajustar límite y para mejor visualización

fig, axes = plt.subplots(2, 2, figsize=(16, 8))
axes = axes.flatten()

plot_calificacion_distribution(tienda, 'Tienda 1', axes[0])
plot_calificacion_distribution(tienda2, 'Tienda 2', axes[1])
plot_calificacion_distribution(tienda3, 'Tienda 3', axes[2])
plot_calificacion_distribution(tienda4, 'Tienda 4', axes[3])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_most_least_sold_products(df, store_name, ax):
    product_counts = df['Producto'].value_counts()

    # Get the most sold product
    most_sold_product = product_counts.head(1).index[0]
    most_sold_count = product_counts.head(1).values[0]

    # Get the least sold product
    least_sold_product = product_counts.tail(1).index[0]
    least_sold_count = product_counts.tail(1).values[0]

    # Create a DataFrame for plotting
    plot_df = pd.DataFrame({
        'Product': [most_sold_product, least_sold_product],
        'Count': [most_sold_count, least_sold_count],
        'Type': ['Más Vendido', 'Menos Vendido']
    })

    sns.barplot(x='Product', y='Count', hue='Type', data=plot_df, palette={'Más Vendido': 'green', 'Menos Vendido': 'red'}, ax=ax)
    ax.set_title(f'Productos Más y Menos Vendidos - {store_name}')
    ax.set_xlabel('Producto')
    ax.set_ylabel('Cantidad Vendida')
    ax.tick_params(axis='x', rotation=45)
    ax.legend(title='Tipo')

fig, axes = plt.subplots(2, 2, figsize=(16, 8))
axes = axes.flatten()

plot_most_least_sold_products(tienda, 'Tienda 1', axes[0])
plot_most_least_sold_products(tienda2, 'Tienda 2', axes[1])
plot_most_least_sold_products(tienda3, 'Tienda 3', axes[2])
plot_most_least_sold_products(tienda4, 'Tienda 4', axes[3])

plt.tight_layout()
plt.show()
```
